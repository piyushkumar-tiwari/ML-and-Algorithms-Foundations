# Differential Calculus for Scalars, Vectors, Matrices, and Deep Learning

---

## Scalar‑valued functions on vector spaces

We work with a function that takes a vector and returns a single number:

$$
f : \mathbb{R}^n \rightarrow \mathbb{R}
$$

The input vector is

$$
\mathbf{x}
=
\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_n
\end{bmatrix}
\in
\mathbb{R}^{n \times 1}
$$

---

### Notation

| Symbol | Meaning |
|--------|---------|
| $n$ | number of input variables |
| $x_i$ | the $i$-th variable |
| $i$ | index, $1 \le i \le n$ |
| $\mathbf{x}$ | column vector of variables |
| $\mathbf{x}^T$ | row‑vector transpose of $\mathbf{x}$ |

---

### Function definition

The function depends on all variables:

$$
f(\mathbf{x}) = f(x_1, x_2, \dots, x_n)
$$

Its output is a scalar:

$$
f(\mathbf{x}) \in \mathbb{R}
$$

---

### Partial derivatives

The partial derivative with respect to $x_i$ is

$$
\frac{\partial f}{\partial x_i}
$$

Each partial derivative is also a scalar:

$$
\frac{\partial f}{\partial x_i} \in \mathbb{R}
$$

Collecting them gives

$$
\left\{
\frac{\partial f}{\partial x_1},
\frac{\partial f}{\partial x_2},
\dots,
\frac{\partial f}{\partial x_n}
\right\}
$$

---

### Total derivative

The total differential is

$$
df = \sum_{i=1}^{n} \frac{\partial f}{\partial x_i} \, dx_i
$$

We can also write it as

$$
df = \nabla f(\mathbf{x})^T \, d\mathbf{x}
$$

where

$$
\nabla f(\mathbf{x})^T \in \mathbb{R}^{1 \times n}, \qquad
d\mathbf{x} \in \mathbb{R}^{n \times 1}
$$

The product $(1\times n)\cdot(n\times 1)$ gives a $1\times 1$ scalar, so $df\in\mathbb{R}$.

---

### Gradient

The gradient is the vector of first‑order partial derivatives:

$$
\nabla f(\mathbf{x})
=
\begin{bmatrix}
\frac{\partial f}{\partial x_1} \\[2pt]
\frac{\partial f}{\partial x_2} \\
\vdots \\
\frac{\partial f}{\partial x_n}
\end{bmatrix}
\in \mathbb{R}^{n \times 1}
$$

Its transpose is $\nabla f(\mathbf{x})^T \in \mathbb{R}^{1 \times n}$.

---

### Hessian matrix

The Hessian contains all second‑order partial derivatives:

$$
H_f(\mathbf{x})
=
\begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} &
\frac{\partial^2 f}{\partial x_1 \partial x_2} &
\cdots &
\frac{\partial^2 f}{\partial x_1 \partial x_n} \\[2pt]
\frac{\partial^2 f}{\partial x_2 \partial x_1} &
\frac{\partial^2 f}{\partial x_2^2} &
\cdots &
\frac{\partial^2 f}{\partial x_2 \partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} &
\frac{\partial^2 f}{\partial x_n \partial x_2} &
\cdots &
\frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}
\in \mathbb{R}^{n \times n}
$$

Each entry is

$$
H_{ij} = \frac{\partial^2 f}{\partial x_i \partial x_j}
$$

where $i$ is the row index and $j$ the column index.

---

### Inverse Hessian

If the Hessian is invertible ($\det(H_f(\mathbf{x})) \neq 0$), then

$$
H_f(\mathbf{x})^{-1} \in \mathbb{R}^{n \times n}
$$

It appears in Newton’s method:

$$
\mathbf{x}_{k+1} = \mathbf{x}_k - H_f(\mathbf{x}_k)^{-1} \, \nabla f(\mathbf{x}_k)
$$

The dimensions work: $(n\times n)\cdot(n\times 1) = (n\times 1)$.

---

## Vector‑valued functions

Now consider a function that returns a vector:

$$
\mathbf{F} : \mathbb{R}^n \rightarrow \mathbb{R}^m
$$

Explicitly,

$$
\mathbf{F}(\mathbf{x})
=
\begin{bmatrix}
f_1(\mathbf{x}) \\
f_2(\mathbf{x}) \\
\vdots \\
f_m(\mathbf{x})
\end{bmatrix}
$$

---

### Notation

| Symbol | Meaning |
|--------|---------|
| $n$ | number of input variables |
| $m$ | number of output components |
| $f_i$ | the $i$-th scalar output function |
| $i$ | index of the output function |
| $j$ | index of the input variable |

The output vector satisfies $\mathbf{F}(\mathbf{x}) \in \mathbb{R}^{m \times 1}$.

---

### Jacobian matrix

The Jacobian generalises the gradient to vector‑valued functions:

$$
J_{\mathbf{F}}(\mathbf{x})
=
\begin{bmatrix}
\frac{\partial f_1}{\partial x_1} &
\frac{\partial f_1}{\partial x_2} &
\cdots &
\frac{\partial f_1}{\partial x_n} \\[2pt]
\frac{\partial f_2}{\partial x_1} &
\frac{\partial f_2}{\partial x_2} &
\cdots &
\frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} &
\frac{\partial f_m}{\partial x_2} &
\cdots &
\frac{\partial f_m}{\partial x_n}
\end{bmatrix}
\in \mathbb{R}^{m \times n}
$$

Each entry is

$$
J_{ij} = \frac{\partial f_i}{\partial x_j}
$$

where $i$ is the row index (output component) and $j$ the column index (input variable).

---

### Total derivative for vector‑valued functions

The total differential becomes

$$
d\mathbf{F} = J_{\mathbf{F}}(\mathbf{x}) \, d\mathbf{x}
$$

Here $J_{\mathbf{F}}(\mathbf{x}) \in \mathbb{R}^{m \times n}$ and $d\mathbf{x} \in \mathbb{R}^{n \times 1}$, so the product is $(m\times n)\cdot(n\times 1) = (m\times 1)$, matching $d\mathbf{F}\in\mathbb{R}^{m\times 1}$.

---

### Hessians for vector‑valued functions

A vector‑valued function does not have a single ordinary matrix as its Hessian. Instead, each scalar component $f_i$ has its own Hessian matrix.

For component $f_i: \mathbb{R}^n \to \mathbb{R}$,

$$
H_{f_i}(\mathbf{x})
=
\begin{bmatrix}
\frac{\partial^2 f_i}{\partial x_1^2} &
\frac{\partial^2 f_i}{\partial x_1 \partial x_2} &
\cdots &
\frac{\partial^2 f_i}{\partial x_1 \partial x_n} \\[2pt]
\frac{\partial^2 f_i}{\partial x_2 \partial x_1} &
\frac{\partial^2 f_i}{\partial x_2^2} &
\cdots &
\frac{\partial^2 f_i}{\partial x_2 \partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f_i}{\partial x_n \partial x_1} &
\frac{\partial^2 f_i}{\partial x_n \partial x_2} &
\cdots &
\frac{\partial^2 f_i}{\partial x_n^2}
\end{bmatrix}
\in \mathbb{R}^{n \times n}
$$

Because there are $m$ component functions, we get $m$ separate Hessian matrices. Together they form a rank‑3 tensor:

$$
\mathcal{H}_{\mathbf{F}}(\mathbf{x}) = \left\{ H_{f_1}(\mathbf{x}), H_{f_2}(\mathbf{x}), \dots, H_{f_m}(\mathbf{x}) \right\}
\in \mathbb{R}^{m \times n \times n}
$$

Each entry is

$$
\mathcal{H}_{ijk} = \frac{\partial^2 f_i}{\partial x_j \partial x_k}
$$

where $i$ indexes the component function, $j$ the first derivative variable, and $k$ the second derivative variable.

---

### Geometric meaning

- Each row of the Jacobian gives the first‑order variation of one component function.
- Each Hessian layer gives the second‑order curvature of that same component.
- The full Hessian tensor stores second‑order information for the whole vector‑valued map.

---

## Matrix‑valued functions

Now let the output be a matrix:

$$
\mathbf{M} : \mathbb{R}^n \rightarrow \mathbb{R}^{m \times p}
$$

Explicitly,

$$
\mathbf{M}(\mathbf{x})
=
\begin{bmatrix}
m_{11}(\mathbf{x}) & m_{12}(\mathbf{x}) & \cdots & m_{1p}(\mathbf{x}) \\
m_{21}(\mathbf{x}) & m_{22}(\mathbf{x}) & \cdots & m_{2p}(\mathbf{x}) \\
\vdots & \vdots & \ddots & \vdots \\
m_{m1}(\mathbf{x}) & m_{m2}(\mathbf{x}) & \cdots & m_{mp}(\mathbf{x})
\end{bmatrix}
$$

---

### Notation

| Symbol | Meaning |
|--------|---------|
| $m$ | number of matrix rows |
| $p$ | number of matrix columns |
| $m_{ij}$ | entry at row $i$, column $j$ |
| $i$ | matrix row index |
| $j$ | matrix column index |
| $k$ | input variable index |

We have $\mathbf{M}(\mathbf{x}) \in \mathbb{R}^{m \times p}$.

---

### Matrix partial derivatives

The derivative with respect to $x_k$ is another matrix:

$$
\frac{\partial \mathbf{M}}{\partial x_k} \in \mathbb{R}^{m \times p}
$$

Its entries are $\frac{\partial m_{ij}}{\partial x_k}$.

---

### Tensor‑valued Jacobian

The Jacobian becomes a rank‑3 tensor:

$$
\frac{\partial \mathbf{M}}{\partial \mathbf{x}} \in \mathbb{R}^{m \times p \times n}
$$

Each entry is $\frac{\partial m_{ij}}{\partial x_k}$.

---

### Tensor‑valued Hessian

The second derivative is a rank‑4 tensor:

$$
\frac{\partial^2 \mathbf{M}}{\partial \mathbf{x}^2} \in \mathbb{R}^{m \times p \times n \times n}
$$

Each entry is $\frac{\partial^2 m_{ij}}{\partial x_k \partial x_l}$, where $i$ is the matrix row, $j$ the column, $k$ the first derivative variable, and $l$ the second derivative variable.

---

## Deep learning uses all three types of functions at different levels.

### 1. Scalar‑valued functions – loss functions

Training a neural network minimises a scalar objective:


L(\mathbf{w}) : \mathbb{R}^n \to \mathbb{R}

Here $\mathbf{w}$ is the parameter vector (size $n$) and $L$ is the loss (e.g., mean squared error, cross‑entropy, policy loss, KL divergence).

Because the output is a scalar, the gradient $\nabla_{\mathbf{w}}L(\mathbf{w}) \in \mathbb{R}^{n \times 1}$ exists. This gradient drives SGD, Adam, RMSProp, momentum, and backpropagation.

Because the loss is scalar, the Hessian $H_L(\mathbf{w}) \in \mathbb{R}^{n \times n}$ also exists. Hessians appear in Newton’s method, curvature analysis, and studies of optimisation stability.

### 2. Vector‑valued functions – neural network layers

A neural network layer is fundamentally a map

$$
\mathbf{F} : \mathbb{R}^n \to \mathbb{R}^m
$$

For example, $\mathbf{h} = \sigma(W\mathbf{x}+\mathbf{b})$ with $\mathbf{x}\in\mathbb{R}^{n\times 1}$, $W\in\mathbb{R}^{m\times n}$, $\mathbf{b}\in\mathbb{R}^{m\times 1}$, $\mathbf{h}\in\mathbb{R}^{m\times 1}$.

Backpropagation repeatedly applies the chain rule through such vector‑valued maps. The key object is the Jacobian matrix $\frac{\partial \mathbf{h}}{\partial \mathbf{x}} = J_{\mathbf{F}}(\mathbf{x})$.

If $\mathbf{y} = f(g(h(\mathbf{x})))$, then

$$
\frac{\partial \mathbf{y}}{\partial \mathbf{x}} = \frac{\partial \mathbf{y}}{\partial g} \frac{\partial g}{\partial h} \frac{\partial h}{\partial \mathbf{x}}
$$

Each factor is a Jacobian. This is the core of backpropagation.

### 3. Matrix‑ and tensor‑valued functions – modern architectures

Matrix‑ and tensor‑valued functions appear in:

- Transformer attention maps,
- CNN feature maps,
- embedding tables,
- batch processing,
- multi‑head attention,
- weight tensors.

Example – attention: $A = \mathrm{softmax}(QK^T)V$ is fundamentally matrix‑to‑matrix computation.

Linear layers use matrix‑valued parameters $W\in\mathbb{R}^{m\times n}$. Convolution kernels are rank‑4 tensors. Transformer activations are rank‑3 tensors.

Full tensor Jacobians and Hessians are rarely computed explicitly because they grow too large. Instead, frameworks like PyTorch, JAX, and TensorFlow compute vector‑Jacobian products (VJP) and Jacobian‑vector products (JVP) implicitly, enabling efficient automatic differentiation.

# The Chain Rule with Jacobians and Backpropagation

## Chain rule for vector-valued functions

Consider the composition

$$
\mathbf{y} = \mathbf{F}(\mathbf{u}), \quad \mathbf{u} = \mathbf{G}(\mathbf{x})
$$

- $\frac{\partial \mathbf{F}}{\partial \mathbf{u}}$ is the Jacobian of $\mathbf{F}$ with respect to its input $\mathbf{u}$.

  Size: $m \times p$ (output dimension of $\mathbf{F}$ × input dimension of $\mathbf{F}$).

  Entries:

  $$
  \frac{\partial F_i}{\partial u_j}
  $$

- $\frac{\partial \mathbf{G}}{\partial \mathbf{x}}$ is the Jacobian of $\mathbf{G}$ with respect to $\mathbf{x}$.

  Size: $p \times n$ (output dimension of $\mathbf{G}$ × input dimension of $\mathbf{G}$).

  Entries:

  $$
  \frac{\partial G_k}{\partial x_l}
  $$

The chain rule for Jacobians states:

$$
\frac{\partial \mathbf{y}}{\partial \mathbf{x}}
=
\left(
\frac{\partial \mathbf{F}}{\partial \mathbf{u}}
\right)
\left(
\frac{\partial \mathbf{G}}{\partial \mathbf{x}}
\right)
$$

This is ordinary matrix multiplication of an $m \times p$ matrix by a $p \times n$ matrix, giving an $m \times n$ Jacobian for the composition.

For longer chains, for example

$$
\mathbf{y} = \mathbf{F}(\mathbf{v}),
\quad
\mathbf{v} = \mathbf{H}(\mathbf{u}),
\quad
\mathbf{u} = \mathbf{G}(\mathbf{x})
$$

we have

$$
\frac{\partial \mathbf{y}}{\partial \mathbf{x}}
=
\left(
\frac{\partial \mathbf{F}}{\partial \mathbf{v}}
\right)
\left(
\frac{\partial \mathbf{H}}{\partial \mathbf{u}}
\right)
\left(
\frac{\partial \mathbf{G}}{\partial \mathbf{x}}
\right)
$$

Each factor is a Jacobian matrix. The dimensions always match because the output dimension of one function equals the input dimension of the next.

---

## Jacobians to backpropagation (scalar loss)

In deep learning we do not compute the full Jacobian of the whole network. For a layer with 1000 inputs and 1000 outputs, that Jacobian would have one million entries.

Instead, we compute gradients of a **scalar loss** $L$ with respect to parameters or inputs.

Suppose we have

$$
L = \text{loss}(\mathbf{y}),
\quad
\mathbf{y} = \mathbf{F}(\mathbf{u}),
\quad
\mathbf{u} = \mathbf{G}(\mathbf{x})
$$

We want

$$
\frac{\partial L}{\partial \mathbf{x}}
$$

which is a row vector.

Using the chain rule for Jacobians:

$$
\frac{\partial L}{\partial \mathbf{x}}
=
\frac{\partial L}{\partial \mathbf{y}}
\;
\frac{\partial \mathbf{y}}{\partial \mathbf{u}}
\;
\frac{\partial \mathbf{u}}{\partial \mathbf{x}}
$$

- $\frac{\partial L}{\partial \mathbf{y}}$ is a $1 \times m$ row vector (derivative of the scalar loss with respect to the output of $\mathbf{F}$)

- $\frac{\partial \mathbf{y}}{\partial \mathbf{u}}$ is the Jacobian of $\mathbf{F}$, size $m \times p$

- $\frac{\partial \mathbf{u}}{\partial \mathbf{x}}$ is the Jacobian of $\mathbf{G}$, size $p \times n$

Multiply from left to right:

1. First compute

   $$
   \mathbf{v}
   =
   \frac{\partial L}{\partial \mathbf{y}}
   \cdot
   \frac{\partial \mathbf{y}}{\partial \mathbf{u}}
   $$

   which gives a $1 \times p$ row vector.

2. Then compute

   $$
   \frac{\partial L}{\partial \mathbf{x}}
   =
   \mathbf{v}
   \cdot
   \frac{\partial \mathbf{u}}{\partial \mathbf{x}}
   $$

   which gives a $1 \times n$ row vector.

---

## What backpropagation actually computes

At each layer we have a **row vector** representing the gradient of the loss with respect to that layer’s output.

We multiply that row vector by the **local Jacobian** of the layer.

This multiplication is called a **vector-Jacobian product (VJP)**.

We never form the full Jacobian matrix.

For each layer we implement a function that:

- Takes the incoming gradient vector (size $1 \times m$)
- Uses the layer’s forward operation to compute the VJP efficiently
- Produces:
  - An outgoing gradient vector (size $1 \times p$) to pass backward
  - The gradient with respect to the layer’s parameters

This is why frameworks like PyTorch, TensorFlow, and JAX can handle millions of parameters: they only ever store and multiply vectors by Jacobians, not Jacobians by Jacobians.

---

## Forward and Backward passes

### Forward pass

Apply a sequence of:

- Linear transformations (matrix multiplications by weights)
- Nonlinear activation functions

The final layer’s output is compared to a target vector using a loss function such as:

- Cross-entropy
- Mean squared error (MSE)

The loss is a scalar.

---

### Gradient at the output

From the loss function compute the gradient of the loss with respect to the final layer’s output (or pre-activation).

That gradient is a **row vector** whose length equals the number of neurons in the last layer.

---

### Backpropagation

Multiply that gradient row vector (size $1 \times m$) by the Jacobian matrix of the last layer’s mapping (size $m \times p$).

The result is a new row vector of size $1 \times p$:

$$
\text{incoming gradient}
\times
\text{local Jacobian}
=
\text{new gradient}
$$

This new vector becomes the incoming gradient for the previous layer.

---

### Dimension conversion

Each multiplication transforms the Jacobian’s dimensions into the gradient’s dimensions.

We never build the full Jacobian of the whole network.

We only propagate a row vector backward, multiplying it by each layer’s local Jacobian.

Thus, multiplying the incoming gradient by the Jacobian converts the Jacobian’s shape into the gradient’s shape, moving the gradient one step backward through the network.

That is exactly how automatic differentiation works in modern deep learning frameworks.

---

### Summary table

| Function type | Output dimension | Gradient | Jacobian | Hessian |
|---------------|------------------|----------|----------|---------|
| $f : \mathbb{R}^n \to \mathbb{R}$ | $\mathbb{R}$ | valid $(n\times 1)$ | trivial | valid $(n\times n)$ |
| $\mathbf{F} : \mathbb{R}^n \to \mathbb{R}^m$ | $\mathbb{R}^{m\times 1}$ | invalid | valid $(m\times n)$ | component‑wise tensor $(m\times n\times n)$ |
| $\mathbf{M} : \mathbb{R}^n \to \mathbb{R}^{m\times p}$ | $\mathbb{R}^{m\times p}$ | invalid | tensor $(m\times p\times n)$ | tensor $(m\times p\times n\times n)$ |

Modern deep learning:

$$
\text{vectorised function composition} \;+\; \text{chain rule} \;+\; \text{optimisation}
$$


### Basic Numerical Analyzer

```python
import sympy as sp
import numpy as np

def make_vector_input():
    x, y, z = sp.symbols("x y z")
    variables = [x, y, z]
    F = sp.Matrix([x**2 + y*z, sp.sin(x) + z**2, x*y - z])
    return F, variables

def make_matrix_input():
    x, y = sp.symbols("x y")
    variables = [x, y]
    M = sp.Matrix([[x**2, x*y], [sp.sin(y), y**2]])
    return M, variables

class DifferentialAnalyzer:
    def __init__(self, F, variables):
        self.F = F
        self.vars = variables
        self.n_vars = len(variables)
        self.rows, self.cols = F.shape
        if self.cols == 1:
            self.input_type = "Column Vector"
            self.is_scalar = (self.rows == 1)
        elif self.rows == 1:
            self.input_type = "Row Vector"
            self.is_scalar = False
        else:
            self.input_type = "Matrix"
            self.is_scalar = False

    def partial_derivatives(self):
        partials = {}
        for v in self.vars:
            partials[v] = self.F.applyfunc(lambda e: sp.diff(e, v))
        return partials

    def gradient(self):
        if not self.is_scalar: return None
        expr = self.F[0, 0]
        return sp.Matrix([sp.diff(expr, v) for v in self.vars])

    def jacobian(self):
        if self.input_type == "Matrix": return None
        flat = self.F.T.tolist()[0] if self.rows == 1 else self.F.tolist()
        flat_exprs = [row[0] if isinstance(row, list) else row for row in flat]
        F_col = sp.Matrix(flat_exprs)
        return F_col.jacobian(self.vars)

    def hessian(self):
        if not self.is_scalar: return None
        expr = self.F[0, 0]
        return sp.hessian(expr, self.vars)

    def evaluate_all(self, point):
        subs = dict(zip(self.vars, point))
        results = {}
        partials_sym = self.partial_derivatives()
        results["partial_derivatives"] = {
            str(v): {
                "symbolic": mat,
                "numerical": np.array(mat.subs(subs).tolist(), dtype=float)
            } for v, mat in partials_sym.items()
        }
        grad_sym = self.gradient()
        if grad_sym is not None:
            results["gradient"] = {"symbolic": grad_sym, "numerical": np.array(grad_sym.subs(subs).tolist(), dtype=float).flatten()}
        else:
            results["gradient"] = None
        jac_sym = self.jacobian()
        if jac_sym is not None:
            results["jacobian"] = {"symbolic": jac_sym, "numerical": np.array(jac_sym.subs(subs).tolist(), dtype=float)}
        else:
            results["jacobian"] = None
        hess_sym = self.hessian()
        if hess_sym is not None:
            results["hessian"] = {"symbolic": hess_sym, "numerical": np.array(hess_sym.subs(subs).tolist(), dtype=float)}
        else:
            results["hessian"] = None
        return results

def print_header(title):
    print("\n" + 25 * "<->" + f"\n{title}\n")

def print_section(label, symbolic, numerical):
    print(f"\n{label}\nSymbolic:\n{sp.pretty(symbolic)}\n\n  Numerical:\n{numerical}")

def print_unavailable(label):
    print(f"\n{label}\nN/A  (not defined for this input type)")

def report(F, variables, point, label):
    analyzer = DifferentialAnalyzer(F, variables)
    results = analyzer.evaluate_all(point)
    print_header(f"ANALYSIS: {label}")
    print(f"\n  Input Function:\n{sp.pretty(F)}\n\n  Type       : {analyzer.input_type}\n  Shape      : {analyzer.rows} x {analyzer.cols}\n  Variables  : {[str(v) for v in variables]}\n  Eval Point : {point}")
    
    print("\nPARTIAL DERIVATIVES\n")
    for var_str, data in results["partial_derivatives"].items():
        print_section(f"∂F/∂{var_str}", data["symbolic"], data["numerical"])
        
    print("\nGRADIENT\n")
    if results["gradient"] is not None:
        print_section("∇f", results["gradient"]["symbolic"], results["gradient"]["numerical"])
    else:
        print_unavailable("∇f  — requires scalar-valued input (R^n -> R)")
        
    print("\nJACOBIAN\n")
    if Henderson_results := results["jacobian"]:
        print_section("J", Henderson_results["symbolic"], Henderson_results["numerical"])
    else:
        print_unavailable("J  — not defined for matrix-valued inputs")
        
    print("\nHESSIAN\n")
    if results["hessian"] is not None:
        print_section("H", results["hessian"]["symbolic"], results["hessian"]["numerical"])
    else:
        print_unavailable("H  — requires scalar-valued input (R^n -> R)")

if __name__ == "__main__":
    x, y, z = sp.symbols("x y z")
    
    report(F=sp.Matrix([x**2 + sp.sin(x*y)]), variables=[x, y], point=[1.0, 0.5], label="Scalar Function  f: R^2 -> R")
    report(F=sp.Matrix([x**2 + y*z, sp.sin(x) + z**2, x*y - z]), variables=[x, y, z], point=[1.0, 2.0, 0.5], label="Vector Function  F: R^3 -> R^3")
    report(F=sp.Matrix([[x**2, x*y], [sp.sin(y), y**2]]), variables=[x, y], point=[1.0, 1.0], label="Matrix Input  M: R^2 -> R^(2x2)")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

class ComputationalNode:
    def __init__(self, expression, variables):
        self.expr = expression
        self.vars = variables
        self.parents = []
        self.local_derivatives = {}
        self.grad_sym = sp.Matrix([sp.diff(expression, v) for v in variables])
        self.jacobian_sym = self.grad_sym.T
        self.hessian_sym = sp.hessian(expression, variables)

    def get_numerical_gradient(self, point):
        grad_funcs = [sp.lambdify(self.vars, g, modules=["numpy"]) for g in self.grad_sym]
        grads = [float(f(*point)) for f in grad_funcs]
        return np.array(grads, dtype=float)

    def get_numerical_hessian(self, point):
        hess_func = sp.lambdify(self.vars, self.hessian_sym, modules=["numpy"])
        H = hess_func(*point)
        return np.array(H, dtype=float)


class VisualizationEngine:
    @staticmethod
    def create_meshgrid(bounds=(-3, 3), resolution=120):
        x_vals = np.linspace(bounds[0], bounds[1], resolution)
        y_vals = np.linspace(bounds[0], bounds[1], resolution)
        X, Y = np.meshgrid(x_vals, y_vals)
        return X, Y

    @staticmethod
    def evaluate_expression(expression, variables, X, Y):
        func = sp.lambdify(variables, expression, modules=["numpy"])
        Z = func(X, Y)
        return np.array(Z, dtype=float)

    @staticmethod
    def evaluate_gradient(grad_sym, variables, X, Y):
        grad_funcs = [sp.lambdify(variables, g, modules=["numpy"]) for g in grad_sym]
        U = grad_funcs[0](X, Y)
        V = grad_funcs[1](X, Y)
        return np.array(U, dtype=float), np.array(V, dtype=float)

    @staticmethod
    def classify_curvature(eigenvalues):
        if np.all(eigenvalues > 0): return "Strictly Convex"
        elif np.all(eigenvalues < 0): return "Strictly Concave"
        elif np.any(eigenvalues > 0) and np.any(eigenvalues < 0): return "Saddle Point"
        else: return "Flat / Degenerate"

    @staticmethod
    def plot_surface(X, Y, Z, title):
        fig = plt.figure(figsize=(8, 6))
        ax = fig.add_subplot(111, projection="3d")
        surf = ax.plot_surface(X, Y, Z, cmap="plasma", alpha=0.85)
        ax.set_title(title); ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("f(x,y)")
        fig.colorbar(surf, shrink=0.6)
        plt.show()

    @staticmethod
    def plot_contours_and_paths(X, Y, Z, U, V, paths, labels, title="Optimization Flow"):
        fig, ax = plt.subplots(figsize=(9, 7))
        contour = ax.contour(X, Y, Z, levels=35, cmap="viridis")
        ax.clabel(contour, inline=True, fontsize=7)
        ax.streamplot(X, Y, U, V, density=1.0, linewidth=1, arrowsize=1)
        colors = ["red", "orange", "cyan"]
        for path, label, color in zip(paths, labels, colors):
            path = np.array(path)
            ax.plot(path[:, 0], path[:, 1], '.-', color=color, linewidth=2, markersize=6, label=label)
            ax.scatter(path[-1, 0], path[-1, 1], color=color, marker='x', s=120)
        ax.set_title(title); ax.set_xlabel("x"); ax.set_ylabel("y"); ax.grid(True); ax.legend()
        plt.show()

    @staticmethod
    def plot_hessian_deformation(hessian, point=(0, 0)):
        square = np.array([[0, 1, 1, 0, 0], [0, 0, 1, 1, 0]])
        transformed = hessian @ square
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.plot(square[0] + point[0], square[1] + point[1], 'b--', linewidth=2, label="Original")
        ax.plot(transformed[0] + point[0], transformed[1] + point[1], 'r-', linewidth=2, label="Hessian Transformation")
        ax.scatter(point[0], point[1], color="black", s=60)
        ax.set_title("Local Curvature Deformation"); ax.grid(True); ax.axis("equal"); ax.legend()
        plt.show()


class OptimizationEngine:
    @staticmethod
    def gradient_descent(node, start_point, lr=0.03, T=60):
        curr = np.array(start_point, dtype=float)
        path = [curr.copy()]
        for _ in range(T):
            grad = node.get_numerical_gradient(curr)
            curr = curr - lr * grad
            path.append(curr.copy())
        return np.array(path)

    @staticmethod
    def momentum_gd(node, start_point, lr=0.03, beta=0.9, T=60):
        curr = np.array(start_point, dtype=float)
        velocity = np.zeros_like(curr)
        path = [curr.copy()]
        for _ in range(T):
            grad = node.get_numerical_gradient(curr)
            velocity = beta * velocity + (1 - beta) * grad
            curr = curr - lr * velocity
            path.append(curr.copy())
        return np.array(path)

    @staticmethod
    def adam(node, start_point, lr=0.08, beta1=0.9, beta2=0.999, epsilon=1e-8, T=60):
        curr = np.array(start_point, dtype=float)
        m, v = np.zeros_like(curr), np.zeros_like(curr)
        path = [curr.copy()]
        for t in range(1, T + 1):
            grad = node.get_numerical_gradient(curr)
            m = beta1 * m + (1 - beta1) * grad
            v = beta2 * v + (1 - beta2) * (grad**2)
            m_hat = m / (1 - beta1**t)
            v_hat = v / (1 - beta2**t)
            curr = curr - (lr * m_hat / (np.sqrt(v_hat) + epsilon))
            path.append(curr.copy())
        return np.array(path)


class VisualizerDirector:
    def __init__(self):
        self.x, self.y = sp.symbols("x y")
        self.vars = [self.x, self.y]

    def analyze(self, expression, name="Function", eval_point=(-1.0, 1.0)):
        print(f"\nAnalyzing: {name}\n")
        node = ComputationalNode(expression, self.vars)
        H = node.get_numerical_hessian(eval_point)
        eigenvalues = np.linalg.eigvals(H)
        curvature = VisualizationEngine.classify_curvature(eigenvalues)

        print("\nGradient:\n", node.grad_sym, "\nJacobian:\n", node.jacobian_sym)
        print("\nHessian:\n", node.hessian_sym, "\nEigenvalues:\n", eigenvalues, "\nCurvature Type:\n", curvature)

        X, Y = VisualizationEngine.create_meshgrid()
        Z = VisualizationEngine.evaluate_expression(expression, self.vars, X, Y)
        U, V = VisualizationEngine.evaluate_gradient(node.grad_sym, self.vars, X, Y)

        gd_path = OptimizationEngine.gradient_descent(node, eval_point)
        momentum_path = OptimizationEngine.momentum_gd(node, eval_point)
        adam_path = OptimizationEngine.adam(node, eval_point)

        VisualizationEngine.plot_surface(X, Y, Z, title=f"{name} Surface")
        VisualizationEngine.plot_contours_and_paths(X, Y, Z, U, V, paths=[gd_path, momentum_path, adam_path], labels=["Gradient Descent", "Momentum GD", "Adam"], title=f"{name} Optimization Dynamics")
        VisualizationEngine.plot_hessian_deformation(H, point=eval_point)


if __name__ == "__main__":
    director = VisualizerDirector()
    x, y = director.x, director.y

    director.analyze(x**2 - y**2, name="Saddle Point", eval_point=(1.0, 1.0))
    director.analyze((10 * (y - x**2)**2 + (1 - x)**2), name="Rosenbrock Landscape", eval_point=(-1.0, 1.0))
    director.analyze(sp.sin(x) + sp.cos(y), name="Trigonometric Landscape", eval_point=(0.5, 0.5))
```
