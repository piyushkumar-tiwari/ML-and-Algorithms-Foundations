### Problem 1 - Deep

In deep.py, implement a program that prompts the user for the answer to the Great Question of Life, the Universe and Everything, outputting Yes if the user inputs 42 or (case-insensitively) forty-two or forty two. Otherwise output No.

```python
def ip():
    ip = input("What is the answer to the Great Question of Life, the Universe and Everything?")
    return (ip.lower()).strip()

def answer(answer):
    if answer == '42':
        print('Yes')
    elif answer == 'forty-two':
        print('Yes')
    elif answer == 'forty two':
        print('Yes')
    else:
        print('No')
    return None

def main():
    inpt = ip()
    ans = answer(inpt)
    return ans

if __name__=='__main__':
    main()
```

### Problem 2 - Bank

In a file called bank.py, implement a program that prompts the user for a greeting. If the greeting starts with “hello”, output $0. If the greeting starts with an “h” (but not “hello”), output $20. Otherwise, output $100. Ignore any leading whitespace in the user’s greeting, and treat the user’s greeting case-insensitively.

```python
def ip():
    inpt = input("Kindly greet me please! :)")
    return (inpt.lower()).strip()

def letter(string):
    cost = 0

    if not string:
        return 100

    if string[0] == 'h':
        if ((string.split())[0]).replace(',', '') == 'hello':
            cost = 0
        else:
            cost = 20
    else:
        cost = 100
    return cost

def main():
    inpt = ip()
    output = letter(inpt)
    print (output,'$')
    

if __name__=='__main__':
    main()
```

.gif, .jpg, .jpeg, .png,

.pdf, .txt, .zip

If the file’s name ends with some other suffix or has no suffix at all, output application/octet-stream instead, which is a common default.

```python
def ip():
    inpt = input('What is your file name? (Please enter alongwith the file extension)')
    return (inpt.lower()).strip()

def filenames(name):
    if name.endswith('.gif'):
        filename = '.gif'
        mediatype = 'media'
    elif name.endswith('.jpg'):
        filename='.jpg'
        mediatype = 'media'
    elif name.endswith('.jpeg'):
        filename='.jpeg'
        mediatype = 'media'
    elif name.endswith('.png'):
        filename= '.png'
        mediatype = 'media'
    elif name.endswith('.pdf'):
        filename= '.pdf'
        mediatype = 'application'
    elif name.endswith('.txt'):
        filename='.txt'
        mediatype = 'application'
    elif name.endswith('.zip'):
        filename = '.zip'
        mediatype = 'application'
    else:
        filename = False
        mediatype = False
    
    return filename, mediatype

def main():
    inpt = ip()
    name, mediatype = filenames(inpt)
    print (name,'->',mediatype)

if __name__=='__main__':
    main()
```
