```python
import numpy as np

def f(variables):
    x = variables[0]
    y = variables[1]
    z = variables[2]
    return 3 * (x**2) + np.sqrt(5 * y) + z

n = [1, 1, 2]
```

### Basic Jacobian Calculation

```python
def jacobian(f, n):
    h = 1e-6
    n = np.asarray(n, dtype=float)
    n_input = np.size(n)
    value = f(n)
    n_output = np.size(value)

    j = np.zeros((n_output, n_input))

    for i in range(n_input):
        n0 = n.copy()
        n1 = n.copy()
        n0[i] += h
        n1[i] -= h
        j[:, i] = (f(n0) - f(n1)) / (2*h)
    return j.T
```

```python
result = jacobian(f, n)
print(result)
```

### Hessian Matrix

```python
def Hessian(f, n):
    h = 1e-6
    na = np.asarray(n, dtype=float)
    f_value = f(n)
    n_input = np.size(na)
    n_output = np.size(f_value)
    H = np.zeros((n_input, n_input))
    for i in range(n_input):
        dell = np.zeros(n_input)
        n0 = n.copy()
        n1 = n.copy()
        n0[i] += h
        n1[i] -= h
        for j in range(n_input):
            dell00 = n0.copy()
            dell01 = n0.copy()
            dell10 = n1.copy()
            dell11 = n1.copy()
            dell00[j] += h
            dell01[j] -= h
            dell10[j] += h
            dell11[j] -= h

            g0 = (f(dell00) - f(dell01))/(2*h)
            g1 = (f(dell10) - f(dell11))/(2*h)
            H[i][j] = (g0-g1)/(2*h)
    return H
```

```python
result = Hessian(f, n)
print(result)
```
