```python
import numpy as np

def f(variables):
    x = variables[0]
    y = variables[1]
    z = variables[2]
    return 3 * (x**2) + np.sqrt(5 * y) + z

n = [1, 1, 2]
```

### Basic Jacobian Calculation

```python
def jacobian(f, n):
    h = 1e-6
    n = np.asarray(n, dtype=float)
    n_input = np.size(n)
    value = f(n)
    n_output = np.size(value)

    j = np.zeros((n_output, n_input))

    for i in range(n_input):
        n0 = n.copy()
        n1 = n.copy()
        n0[i] += h
        n1[i] -= h
        j[:, i] = (f(n0) - f(n1)) / (2*h)
    return j.T
```

```python
result = jacobian(f, n)
print(result)
```
