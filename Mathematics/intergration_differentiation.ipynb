### Reimann Integration Method

```python

import math
import numpy as np

def choose_your_function():
    k = input('Please enter the function you want to integrate, its upper limit, and lower limit respectively seperated by the comma')
    k = k.split(',')
    k1 = float(k[1])
    k2 = float(k[2])
    lst = {
        'Square': lambda x: x**2,
        'Sin': np.sin,
        'Exp': np.exp,
        'Root': np.sqrt,
        'Tan': np.tan,
        'Cos': np.cos
    }
    func_name = k[0]
    func = lst[func_name]
    print(func_name, k1, k2)
    return func, k1, k2

def integration(f, up_lim, low_lim):
    delta = 0.00001
    times = int((abs(up_lim - low_lim))/ delta)
    x = low_lim
    y = 0
    for i in range(times):
        num = f(x)
        num = num * delta
        y = y+num
        x = x+delta
    return y

fun, k1, k2 = choose_your_function()
print(integration(fun, k1, k2))
```
